```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#lets set the  parameters that control the general style of the plot like color of the background and whether the grid id enabled by default or not
sns.set_style('darkgrid')
#rcParams = runtime configuration parameters
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE5B4' #peach color code
```

```python
data= pd.read_csv('world-happiness-report-2021.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# we will use only coloums we are intrested in
data_col=['Country name','Regional indicator', 'Happiness score','Logged GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
data = data[data_col].copy()
```

```python
data.head()
```

```python
data.info()
```

```python
#Rename the column names
happy_df=data.rename(columns={'Country name':'country_name' , 'Regional indicator':'regional_indicator', 'Happiness score':'happiness_score', 'Logged GDP per capita':'logged_GDP_per_capita', 'Social support':'social_support', 'Healthy life expectancy':'healthy_life_expectancy', 'Freedom to make life choices':'freedom_to_make_life_choices', 'Generosity':'generosity', 'Perceptions of corruption':'perceptions_of_corruption' })
```

```python
happy_df.head()
```

```python
#checking null values
happy_df.isnull().sum()
```

```python
#Visualizations
#Plot b/w Happiness and GDP for diff regions
```

```python
plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot Between Happiness Score and GDP')
sns.scatterplot(x=happy_df.happiness_score, y=happy_df.logged_GDP_per_capita, hue= happy_df.regional_indicator, s=200);

plt.legend(loc = 'upper left', fontsize= '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per Capita')
```

```python
# we can see for Western Europe and North America the hapiness score and GDP are high
#Sub saharan Africa has hapiness score and gdp as low
```

```python
#plot GDP by region by this we will know which region has highest percentage contribution to the worlds GDP as per the data
```

```python
gdp_region = happy_df.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
gdp_region
```

```python
#plot with pie plot

gdp_region.plot.pie(autopct= '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

```python
# we can see sub-Saharan Africa is contributing most to the worlds GDP with 20.7% . The reason for it being Sub-Saharan Africa contains around 36 countries
# Next its Western Europe countries contributing are contributing 16.2% to the world GDP
#North America and ANZ contributes the least with 3.1% to the World GDP, as it consists of 4 countries America, Australia, Canada and Newzeland
```

```python
#Count total number of countries in the region
total_country = happy_df.groupby('regional_indicator')[['country_name']].count()
print(total_country)
```

```python
#Correlation map

cor = happy_df.corr(method= 'pearson')
f,ax= plt.subplots(figsize =(10,5))
sns.heatmap(cor,cmap='Blues',annot = True,mask=np.zeros_like(cor,dtype=np.bool), square=True, ax=ax)
```

```python
# we can see that there is negative corelation between happiness score and perception of corruption. This means if there is lot of curruption in the country then the citizens are less happy
# 'perception of corruption' and 'freedom to make life choices' has negetive corelation
#there is negetive between perception of corruption and all other parameters
#there is negetive between generosity and all other parameters
```

```python
#Visualize a bar plot to see corruption in different regions
#we are using mean() to find the average corruption in each of these regions
corruption = happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

```python
#let's visualize
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of corruption in various Regions')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption Index', fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.bar(corruption.index, corruption.perceptions_of_corruption)
          
```

```python
# As per the table there is least corruption in 'North America and ANZ' and 'Western Europe'
#There is more currution in 'Central and Eastern Europe' and 'South Asia'
```

```python
# Top 10 Happy countries and bottom 10 happy countries
top_10 = happy_df.head(10)
bottom_10= happy_df.tail(10)
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad = 2)

#Top 10 Happy countries
xlabels = top_10.country_name
axes[0].set_title('Top 10 Happiest Countries Life Expectancy')
axes[0].set_xticklabels(xlabels, rotation=45, ha ='right')

sns.barplot(x= top_10.country_name, y=top_10.healthy_life_expectancy, ax=axes[0])

axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life expectancy')


#Bottom 10 happy countries

xlabels = bottom_10.country_name
axes[1].set_title('Bottom 10 Happiest Countries Life Expectancy')
axes[1].set_xticklabels(xlabels, rotation=45, ha ='right')

sns.barplot(x= bottom_10.country_name, y=bottom_10.healthy_life_expectancy, ax=axes[1])

axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life expectancy')
```

```python
# From above we can say that in top 10 happiest countries the average life expectancy is above 70 years on average
# In bottom 10 least happy countries Lesotho has less than 50 years life expectancy and in these countries people would live less than 60 years on average
# so if we compare people living in top 10  and bottom 10 happy countries, top 10 happy countries live 10 years more than bottom 10 countries on average
```

```python
# Let plot between 'freedom to make life choices' and 'happiness score'

plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=happy_df.freedom_to_make_life_choices, y=happy_df.happiness_score, hue= happy_df.regional_indicator, s=200 )
plt.legend(loc = 'upper left', fontsize = '12')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness score')
```

```python
# western europe has highest happiness score and freedom to make life choices
# middle east and North africa and sub-Saharan Africa has lowest score in both freedom to make life choices and happiness score
#southeast asia the free to make life choice is more but less happiness score
```

```python
# Above contries have high corruption index
```
